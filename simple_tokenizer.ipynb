```python
import re
class Vocab:
    def __init__(self, raw_data):
        self.vocab = self.make_vocab(raw_data)

    def token_to_idx(self, sorted_tokens):
        vocabulary = {}
        for idx, token in enumerate(sorted_tokens):
            vocabulary[token] = idx
        return vocabulary


    def make_vocab(self, raw_data):
        raw_list = re.split(r"([_,\":*$#+!()])|\s", raw_data)
        raw = [token.strip() for token in raw_list if token and token.strip()]
        unique_token_sorted = sorted(set(raw))
        return self.token_to_idx(unique_token_sorted)
        
            
    def extend_vocab(self, array_token):
        vocabulary = {}
        array_token = list(set(array_token))
        extend_ = list(self.vocab.keys())
        extend_.extend(array_token)
        sorted(extend_)
        self.vocab = self.token_to_idx(extend_)
        
        
        

    

```

```python
class Tokenizer:
    def __init__(self, vocab):
        self.str_to_int = vocab
        self.int_to_str = {i:w for w,i in vocab.items()}


    
        
    def encoded(self,raw_data):
        idx = []
        raw_list = re.split(r"([_,\":*$#+!()])|\s", raw_data)
        raw = [token.strip() for token in raw_list if token and token.strip()]
        for token in raw:
            if token in self.str_to_int.keys():
                idx.append(self.str_to_int[token])
            else:
                idx.append(self.str_to_int["<|unk|>"])
        return idx


    def decoded(self, idx):
        text = []
        for id in idx:
            text.append(self.int_to_str[id])
        raw_joined_text = " ".join(text)
        decoded_text = re.sub(r"\s+([_–,.:?!#$&@\"'()])",r'\1',raw_joined_text)
        return decoded_text

        
        

        
```

```python
text = Vocab(raw_data)
text.extend_vocab(["<|unk|>", "<|endoftext|>"])
vocab = text.vocab
```
